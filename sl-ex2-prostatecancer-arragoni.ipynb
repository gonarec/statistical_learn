**27. Generate a linear regression model using predictorsTrain_std as dependent variables and lpsaTrain as independent variable**

```python
linear_reg = LinearRegression(fit_intercept=True)

# Fit del modello ai dati di addestramento
linear_reg.fit(predictorsTrain_std2, lpsaTrain)
```

**28. Show the parameters of the linear regression model computed above. Compare the parameters with those shown in Table 3.2 of the book (page 50)**

```python
print('Linear regression parameters:\n ', linear_reg.get_params(),'\n')
coefficients = [linear_reg.intercept_]
coefficients.extend(linear_reg.coef_)
print('Coefficients: \n', coefficients) # Beta
```

```python
N=len(predictorsTrain_std2)
p=len(predictorsTrain_std2.columns)+1
X_with_intercept = np.empty(shape=(N, p), dtype=float)
X_with_intercept[:, 0] = 1
X_with_intercept[:, 1:p] = predictorsTrain_std2.values
view=pd.DataFrame(X_with_intercept)
view
```

```python

y_predict = linear_reg.predict(predictorsTrain_std2) 
residual_sum_of_squares = ((lpsaTrain.values - y_predict)**2).sum()

sigma_squared_hat = residual_sum_of_squares /(N - p) 

vj = np.linalg.inv(X_with_intercept.T @ X_with_intercept)

std_error=[]
for x in range(p):
    std_error.append((sigma_squared_hat ** 0.5) * (vj[x,x] ** 0.5))
    
#calculate the Z score
z_score = []

for x in range(len(coefficients)):
    z_score.append(coefficients[x] / std_error[x])

result=pd.DataFrame({'coeff': coefficients, 
                   'std_err': std_error,
                   'z_score': z_score})
result
```
